# U.S. Medical Insurance Costs

#
Project Scope:

-1.Load dataset into a dictionary so we can access each persons data

-2. Calculate the average age, aswell as the minimum and maximimum age in the dataset

-3.Calculate the average charges for Smokers vs Non-Smokers

-4. Find the number of people from each of the different areas, and the average charges for each of them

```python
import csv

# importing the dataset into dictionary
with open('insurance.csv', mode = 'r') as data:
    csv_reader = csv.DictReader(data)
    medical_records = [row for row in csv_reader]
    
    
# Calculating the average age of the dataset
def get_average_age(medical_records):
    total_age = 0
    min_age = 1000
    max_age = 0
    for i in range(len(medical_records)):
        age = int(medical_records[i]['age'])
        total_age += age
        if age < min_age:
            min_age = age
        if age > max_age:
            max_age = age
    
    return total_age / len(medical_records), min_age, max_age

average_age, min_age, max_age = get_average_age(medical_records)
print("The average age is", str(round(average_age,0)), "years old")
print("The maximum age is", str(max_age))
print("The minimum age is", str(min_age))



# Comparing the average cost for smokers vs non-smokers
def smoking(medical_records):
    non_smoking_total = 0
    smoking_total = 0
    num_non_smoking = 0
    num_smoking = 0
    
    for i in range(len(medical_records)):
        if medical_records[i]["smoker"] == 'yes':
            num_smoking +=1
            smoking_total += float(medical_records[i]['charges'])
        elif medical_records[i]["smoker"] == 'no':
            num_non_smoking += 1
            non_smoking_total += float(medical_records[i]['charges'])
    
    return smoking_total/num_smoking, non_smoking_total / num_non_smoking, num_smoking, num_non_smoking


smoking_avg, non_smoking_avg, num_smoking, num_non_smoking = smoking(medical_records)

print("\nThere are", str(num_smoking),"smokers, and the average cost for smokers is", str(round(smoking_avg, 2)),"dollars, whereas there are ", str(num_non_smoking),"non-smokers and the average cost for non-smokers is", str(round(non_smoking_avg,2)), "dollars \n")


# Calculating the average cost for each area, aswell as the number of people from each one
def find_location(medical_records):
    northeast = 0
    northeast_total = 0
    northwest = 0
    northwest_total = 0
    southeast = 0
    southeast_total = 0
    southwest = 0
    southwest_total = 0
    
    for i in range(len(medical_records)):
        region = medical_records[i]['region']
        charge = float(medical_records[i]['charges'])
        if region == 'northeast':
            northeast += 1
            northeast_total += charge
        elif region == 'northwest':
            northwest += 1
            northwest_total += charge
        elif region == 'southeast':
            southeast += 1
            southeast_total += charge
        elif region == 'southwest':
            southwest += 1
            southwest_total += charge
            
    
    return [northeast, northwest, southeast, southwest], [northeast_total/northeast, northwest_total/northwest, southeast_total/southeast, southwest_total/southwest]



location_tot, location_avg = find_location(medical_records)

print("The number of people from the northeast is", str(location_tot[0]), "with an average cost of", str(round(location_avg[0],2)), "dollars")
print("The number of people from the northwest is", str(location_tot[1]), "with an average cost of", str(round(location_avg[1],2)), "dollars")
print("The number of people from the southeast is", str(location_tot[2]), "with an average cost of", str(round(location_avg[2],2)), "dollars")
print("The number of people from the southwest is", str(location_tot[3]), "with an average cost of", str(round(location_avg[3],2)), "dollars \n")


# Computing the number of males and females in the dataset, aswell as the averge charges for both sexes
def sex_totals(medical_records):
    male = 0
    female = 0
    male_total_charge = 0
    female_total_charge = 0
    
    for i in range(len(medical_records)):
        sex = medical_records[i]['sex']
        charge = float(medical_records[i]['charges'])
        if sex == 'male':
            male += 1
            male_total_charge += charge
        elif sex == 'female':
            female += 1
            female_total_charge += charge
            
    return male, female, male_total_charge / male, female_total_charge / female


males, females, male_avg_charge, female_avg_charge = sex_totals(medical_records)

print(f"There are {males} males and their average charges are {round(male_avg_charge,2)} dollars")
print(f"There are {females} females and their average charges are {round(female_avg_charge,2)} dollars \n")



def analyze_bmi(medical_records):
    underweight = 0
    underweight_charges = 0
    healthy_weight = 0
    healthy_weight_charges = 0
    overweight = 0
    overweight_charges = 0
    obese = 0
    obese_charges = 0
    
    for i in range(len(medical_records)):
        bmi = float(medical_records[i]['bmi'])
        charge = float(medical_records[i]['charges'])
        
        if bmi <= 18.5:
            underweight += 1
            underweight_charges += charge
        elif 18.5 < bmi <= 24.9:
            healthy_weight += 1
            healthy_weight_charges += charge
        elif 24.9 < bmi <= 29.9:
            overweight += 1
            overweight_charges += charge
        elif bmi > 29.9:
            obese += 1
            obese_charges += charge
    
    print(f"There are {underweight} underweight people, and their average charges are {round(underweight_charges/underweight,2)} dollars")
    print(f"There are {healthy_weight} healthy weighted people, and their average charges are {round(healthy_weight_charges/healthy_weight,2)} dollars")
    print(f"There are {overweight} overweight people, and their average charges are {round(overweight_charges/overweight,2)} dollars")
    print(f"There are {obese} obese people, and their average charges are {round(underweight_charges/underweight,2)} dollars \n")
    

    
analyze_bmi(medical_records)


def analyze_num_children(medical_records):
    zero = 0
    zero_charges = 0
    one = 0
    one_charges = 0
    two = 0
    two_charges = 0
    three_or_more = 0
    three_or_more_charges = 0
    
    for i in range(len(medical_records)):
        childs = float(medical_records[i]['children'])
        charge = float(medical_records[i]['charges'])
        
        if childs == 0:
            zero += 1
            zero_charges += charge
        elif childs == 1:
            one += 1
            one_charges += charge
        elif childs == 2:
            two += 1
            two_charges += charge
        elif childs >= 3:
            three_or_more += 1
            three_or_more_charges += charge
    
    print(f"There are {zero} people with no children, and their average charges are {round(zero_charges/zero, 2)} dollars")
    print(f"There are {one} people with one child, and their average charges are {round(one_charges/one, 2)} dollars")
    print(f"There are {two} people with two children, and their average charges are {round(two_charges/two, 2)} dollars")
    print(f"There are {three_or_more} people with three or more children, and their average charges are {round(three_or_more_charges/three_or_more, 2)} dollars")

    



analyze_num_children(medical_records)



```

From this analysis of the data, we can see that people who are :

-Smokers

-From the Southeast

-Male

-Overweight

-Have two children

tend to be charged more on average.
